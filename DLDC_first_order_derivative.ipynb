```python
import numpy as np
from tensorflow.keras.layers import Input, Dense, Layer
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.constraints import min_max_norm
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 300
## DLDC for better approximating the differential operator
## Generate dataset form arbitrary function
x = np.linspace(0, 2*np.pi, 360)
dx = x[1]-x[0]
y = np.sin(x)
dydx = np.gradient(y, dx)
## Re-arraging data
X = np.column_stack((y[1:], y[:-1]))
Y = dydx[1:]
```

```python
## Create model
model = Sequential()
model.add(Dense(units=1, kernel_constraint=min_max_norm(min_value=-57.13, max_value=57.13, rate=1.0, axis=1), activation='linear', use_bias=False))
# Compile and fit
model.compile(optimizer= Adam(learning_rate=1), loss='mse')
history = model.fit(X, Y, epochs=500)
print(X)
# Check weights
w = model.layers[0].get_weights()
print('Wights are '+str(w[0][0])+' and '+str(w[0][1]))
print('1/dx is equal to: '+str(1/dx))
```

```python
# Plot result
DLDC_dydx = model.predict(X)
plt.figure()
plt.plot(x, y, label='$sin(x)$')
plt.plot(x, dydx, label='Analytical dy/dx = cos(x)')
plt.plot(x[1:], DLDC_dydx, label='DLDC $dy/dx$')
plt.xlabel('$x$',fontsize=16)
plt.ylabel('$y=sin(x)$ and $dy/dx$',fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend()
plt.show(block=False)
plt.savefig('sinusoidal_function.png', dpi=300, bbox_inches='tight')
```
